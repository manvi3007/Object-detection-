```python
model = cv2.dnn_DetectionModel(frozen_model, config_file)
print("Model loaded successfully!")
```

```python
classLabels = []
file_name = 'C:/Users/manvi/OneDrive/Desktop/EPICS/labels.txt'
with open (file_name,'rt') as fpt :
    classLabels = fpt.read().rstrip('\n').split('\n')
```

```python
print(classLabels)
```

```python
print(len(classLabels))
```

```python
model.setInputSize(640, 640)
model.setInputScale(1.0 / 127.5)
model.setInputMean((127.5, 127.5, 127.5)) 
model.setInputSwapRB(True)
```

```python
img = cv2.imread('C:/Users/manvi/.ipynb_checkpoints/boy.jpeg')
plt.imshow(img)
```

```python
ClassIndex, confidence, bbox = model.detect(img, confThreshold=0.3)
```

```python
print(ClassIndex)
```

```python
font_scale = 1.5
font = cv2.FONT_HERSHEY_PLAIN

for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
    # Validate the ClassInd to avoid index errors
    if 1 <= ClassInd <= len(classLabels):
        cv2.rectangle(img, boxes, (255, 0, 0), 2)
        cv2.putText(img, classLabels[ClassInd-1], 
                    (boxes[0] + 10, boxes[1] + 40), 
                    font, fontScale=font_scale, 
                    color=(0, 255, 0), thickness=2)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Load the video
cap = cv2.VideoCapture('C:/Users/manvi/.ipynb_checkpoints/video.mp4')
if not cap.isOpened():
    cap = cv2.VideoCapture(0)  # Try the webcam if the video cannot be opened
if not cap.isOpened():
    raise IOError('Cannot open the video')

font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN

while True:
    ret, frame = cap.read()

    if not ret:
        print("Video has ended or cannot be read.")
        break

    # Perform object detection
    ClassIndex, confidence, bbox = model.detect(frame, confThreshold=0.55)

    if len(ClassIndex) != 0:  # If objects are detected
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
            if ClassInd <= 80:  # Assuming 80 is the total number of class labels
                # Draw the bounding box
                cv2.rectangle(frame, boxes, (255, 0, 0), 2)
                # Add the label text
                cv2.putText(frame, classLabels[ClassInd-1], 
                            (boxes[0] + 10, boxes[1] + 40), font, 
                            fontScale=font_scale, color=(0, 255, 0), thickness=2)

    # Display the frame with detection
    cv2.imshow('Object Detection', frame)

    # Exit the loop when 'q' is pressed
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
